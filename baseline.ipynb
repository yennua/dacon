```
from sklearn.model_selection import KFold
```

```
k_fold = KFold(n_splits=5, shuffle=True, random_state=42)

for train_idx, val_idx in k_fold.split(x):
    print(train_idx, val_idx)
```

```
k_fold = KFold(n_splits=5, shuffle=True, random_state=42)

models = []

#loc는 로우의 네임을 인덱스값으로 쓰지만 iloc는 로우의 위치값을 인덱스로 씀
for train_idx, val_idx in k_fold.split(x_train):
    X_train, Y_train = x_train.iloc[train_idx], y_train.iloc[train_idx]
    X_val, Y_val = x_train.iloc[val_idx], y_train.iloc[val_idx]
    
    #라이트데이터셋에 묶어서 넣어야 함 
    d_train = lgb.Dataset(X_train, Y_train)
    d_val = lgb.Dataset(X_val, Y_val)

    params = {
            'learning_rate': 0.01, # 가중치 변화 폭, 학습률
            'objective': 'binary', # regression(회귀), binary(이진분류), multiclass(다중분류)
            'metric':'auc',
            'seed':42 #랜덤값 고정
            }
    
    models.append(lgb.train(params, d_train, 100, d_val, verbose_eval=20, early_stopping_rounds=10))
```

```
preds = []
for model in models:
    preds.append(model.predict(x_test))
len(preds)
```

```
pred = (preds[0]+preds[1]+preds[2]+preds[3]+preds[4])/5
submission['nerdiness'] = pred
submission
```
