```
country_map = {}

for i, c in enumerate(train['country'].unique()):
    country_map[c] = i
```

```
major_map = {}

for i, c in enumerate(train['major'].unique()):
    major_map[c] = i
```

```
train['country'] = train['country'].map(country_map)
test['country'] = test['country'].map(country_map)
```

```
train['major'] = train['major'].map(major_map)
test['major'] = test['major'].map(major_map)
```

```
x_train = train.loc[:, 'Q1':'ASD']
y_train = train['nerdiness']

x_test = test.loc[:, 'Q1':'ASD']
```

```
x_train.head()
```

```
import lightgbm as lgb
```

```
lgb_model = lgb.LGBMClassifier(n_jobs=-1, random_state=42)
```

```
lgb_model.fit(x_train, y_train)
```

```
pred = lgb_model.predict_proba(x_test)[:,1]
```

```
submission['nerdiness'] = pred
```
